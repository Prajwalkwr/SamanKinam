```python
from PIL import Image, ImageDraw, ImageFont
import os
os.makedirs('src/assets', exist_ok=True)
width, height = 1200, 280
img = Image.new('RGB', (width, height), '#FF6B00')
draw = ImageDraw.Draw(img)
for i in range(width):
    r = int(255 - (i / width) * 40)
    g = int(107 + (i / width) * 50)
    b = int(0 + (i / width) * 30)
    draw.line([(i, 0), (i, height)], fill=(r, g, b))
try:
    font_big = ImageFont.truetype('arialbd.ttf', 64)
    font_small = ImageFont.truetype('arial.ttf', 36)
    font_button = ImageFont.truetype('arialbd.ttf', 32)
except Exception:
    font_big = ImageFont.load_default()
    font_small = ImageFont.load_default()
    font_button = ImageFont.load_default()
text1 = '4.4'
text2 = 'नव वर्ष उत्सव'
block_x, block_y = 40, 40
block_w, block_h = 340, 180
draw.rectangle([block_x, block_y, block_x + block_w, block_y + block_h], fill=(255, 255, 255, 200))
draw.text((block_x + 20, block_y + 20), text1, font=font_big, fill='#FF6B00')
draw.text((block_x + 20, block_y + 110), text2, font=font_small, fill='#4B0082')
header = '4.4 MEGA SALE IS LIVE NOW'
w3, h3 = draw.textsize(header, font=font_big)
draw.text(((width - w3) / 2, 70), header, font=font_big, fill='white')
button_text = 'SHOP NOW'
w4, h4 = draw.textsize(button_text, font=font_button)
button_x = width - w4 - 120
button_y = 170
draw.rectangle([button_x - 20, button_y - 10, button_x + w4 + 20, button_y + h4 + 10], fill='white')
draw.text((button_x, button_y), button_text, font=font_button, fill='#FF6B00')
img_path = 'src/assets/banner-mobile.jpg'
img.save(img_path, quality=95)
print(f'Created banner {img_path}')
```
